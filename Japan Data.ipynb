```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the CSV file
Japan_insights = pd.read_csv('data/Japan_Market_Insights.csv', index_col=0)
```

```python
Japan_insights = Japan_insights.iloc[:, :25]

# Step 2: Transpose the DataFrame so that rows become columns
Japan_insights_transposed = Japan_insights.T

# Step 3: Verify the transposed data structure
print(Japan_insights_transposed.head())

# Step 4: Compute the correlation matrix for the transposed DataFrame
correlation_matrix = Japan_insights_transposed.corr()

# Step 5: Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Step 6: Add title and display the heatmap
plt.title('Correlation Heatmap between Tourism Variables in Japan (Transposed)')
plt.show()
Japan_insights_transposed
```

## Showing Yen's Depreciation with Real Exchange Rate

```python
# Load the data
df = pd.read_csv('data/RBJPBIS.csv')
df2 = pd.read_csv('data/RBKRBIS.csv')
df3 = pd.read_csv('data/RBCNBIS.csv')

# Convert the DATE column to datetime
df['DATE'] = pd.to_datetime(df['DATE'])
df2['DATE'] = pd.to_datetime(df2['DATE'])
df3['DATE'] = pd.to_datetime(df3['DATE'])

# Subset the data between 2007 and 2024
subset_df = df[(df['DATE'] >= '2007-01-01') & (df['DATE'] <= '2024-12-31')]
subset_df2 = df2[(df2['DATE'] >= '2007-01-01') & (df2['DATE'] <= '2024-12-31')]
subset_df3 = df3[(df3['DATE'] >= '2007-01-01') & (df3['DATE'] <= '2024-12-31')]

# Create a line chart with customized colors and styles
plt.figure(figsize=(10, 6))

# Japan line with increased thickness and custom color
plt.plot(subset_df['DATE'], subset_df['RBJPBIS'], color='blue', linewidth=2.5, linestyle='-', label='Real Broad Effective Exchange Rate (Japan)')

# China line with a specific color
plt.plot(subset_df3['DATE'], subset_df3['RBCNBIS'], color='red', linewidth=1.5, linestyle='-.', label='Real Broad Effective Exchange Rate (China)')

# Korea line with a specific color
plt.plot(subset_df2['DATE'], subset_df2['RBKRBIS'], color='green', linewidth=1.5, linestyle='--', label='Real Broad Effective Exchange Rate (Korea)')

# Title and labels
plt.title('Japan, Korea, and China Real Broad Effective Exchange Rates (2007-2024)')
plt.xlabel('Year')
plt.ylabel('Real Broad Effective Exchange Rate')

# Grid and legend
plt.grid(True)
plt.legend()

# Source text
plt.text(0.5, -0.15, 'Source: Bank for International Settlements', 
         fontsize=10, ha='center', va='center', transform=plt.gca().transAxes)

# Adjust layout and display
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import folium
import requests

# Load the data (replace the path with your local path)
df = pd.read_csv('data/trips-japan.csv')
df_clean = df.iloc[2:, :2]  # Select rows starting from row 2 and only the first two columns
df_clean.columns = ['Prefecture', 'Overnight_Stays']  # Rename columns
df_clean['Overnight_Stays'] = df_clean['Overnight_Stays'].str.replace(',', '').astype(float)  # Convert stays to float

# Load GeoJSON for Japan prefectures
url = "https://raw.githubusercontent.com/dataofjapan/land/master/japan.geojson"
response = requests.get(url)
japan_geo = response.json()

# Create a base map centered around Japan
m = folium.Map(location=[36.2048, 138.2529], zoom_start=5)

# Add a choropleth layer to the map using the cleaned data
folium.Choropleth(
    geo_data=japan_geo,
    name='choropleth',
    data=df_clean,
    columns=['Prefecture', 'Overnight_Stays'],
    key_on='feature.properties.nam',
    fill_color='BuPu',  # Change color scheme to BuPu for better contrast
    fill_opacity=0.5,   # Increase transparency
    line_opacity=0.8,   # Adjust boundary line opacity
    line_color='white', # Change line color to white for better contrast
    legend_name='Overnight Stays by Prefecture (in thousands)'
).add_to(m)

# Save the map to an HTML file and display
m.save('japan_overnight_stays_map_improved.html')


```
